## Question
Are the trends in immigration from Mexico to the USA visible in the popularity of baby names? 
## Data Sources
1. All baby names since 1880 in the US (source: https://www.ssa.gov/oact/babynames/limits.html)

2. Imigration trends from Mexico to the USA: 

"Immigration from Latin America and the Caribbean started gradually in the 1960s, grew at an increasing rate from 1970 to 2000, and then rose at a decreasing rate from 2006 to 2019, as shown earlier in Figure 1: Foreign-Born Shares of the US Population."
(source: Journal of Economic Perspectives—Volume 37, Number 1—Winter 2023—Pages 199–222, https://pubs.aeaweb.org/doi/pdfplus/10.1257/jep.37.1.199)

![Immigration from Latin America and Caribbean, 1960–2019]('../10_project_presentation/figure_1.png')

3. Most popular names in Mexico throughout the years. (source: 'Estos fueron los nombres de bebé más usados durante el siglo XX', https://www.univision.com/estilo-de-vida/madres/estos-fueron-los-nombres-de-bebe-mas-usados-durante-el-siglo-xx

- Jose
- Carlos
- Alejandro
- Javier
- Juan
- Miguel

```python
import pandas as pd
```

```python
babies = pd.read_csv('../05_methods_functions_attributes/final_df_all_names.csv', index_col = 0)
```

```python
# creating a data frame for each name

# Jose
jose = (babies['name'] == 'Jose') & (babies['sex'] == 'M')
df_jose = babies.loc[jose]

#Carlos
carlos = (babies['name'] == 'Carlos') & (babies['sex'] == 'M')
df_carlos = babies.loc[carlos]

# Alejandro:
alejandro = (babies['name'] == 'Alejandro') & (babies['sex'] == 'M')
df_alejandro = babies.loc[alejandro]

#Javier
javier = (babies['name'] == 'Javier') & (babies['sex'] == 'M')
df_javier = babies.loc[javier]

#Juan
juan = (babies['name'] == 'Juan') & (babies['sex'] == 'M')
df_juan = babies.loc[juan]

#Miguel
miguel = (babies['name'] == 'Miguel') & (babies['sex'] == 'M')
df_miguel = babies.loc[miguel]
```

```python
#now importing matplotlib
import matplotlib.pyplot as plt
```

```python
#creating a plot for each name:

#variables
x = df_jose['year']
y = df_jose['frequency']

x1 = df_carlos['year']
y1 = df_carlos['frequency']

x2 = df_alejandro['year']
y2 = df_alejandro['frequency']

x3 = df_javier['year']
y3 = df_javier['frequency']

x4 = df_juan['year']
y4 = df_juan['frequency']

x5 = df_miguel['year']
y5 = df_miguel['frequency']
```

```python
plt.plot(x,y, color='skyblue', label= 'Jose');
plt.title('Popularity of the name Jose in the USA (1880-2020)')
plt.legend(loc=2)
plt.xlabel('year')
plt.ylabel('frequency');
           
```

```python
plt.plot(x1,y1, color='pink', label= 'Carlos');
plt.title('Popularity of the name Carlos in the USA (1880-2020)')
plt.legend(loc=2)
plt.xlabel('year')
plt.ylabel('frequency');
```

```python
plt.plot(x2,y2, color='green', label= 'Alejandro');
plt.title('Popularity of the name Alejandro in the USA (1880-2020)')
plt.legend(loc=2)
plt.xlabel('year')
plt.ylabel('frequency');
```

```python
plt.plot(x3,y3, color='yellow', label= 'Javier');
plt.title('Popularity of the name Javier in the USA (1880-2020)')
plt.legend(loc=2)
plt.xlabel('year')
plt.ylabel('frequency');
```

```python
#now plotting the data for female names:
plt.plot(x4,y4, color='brown', label= 'Juan');
plt.title('Popularity of the name Juan in the USA (1880-2020)')
plt.legend(loc=2)
plt.xlabel('year')
plt.ylabel('frequency');
```

```python
plt.plot(x5,y5, color='red', label= 'Miguel');
plt.title('Popularity of the name Miguel in the USA (1880-2020)')
plt.legend(loc=2)
plt.xlabel('year')
plt.ylabel('frequency');
```

```python
#now merging these plots into one
plt.plot(x, y, color='skyblue', label='Jose')
plt.plot(x1, y1, color='pink', label='Carlos')
plt.plot(x2, y2, color='green', label='Alejandro')
plt.plot(x3,y3, color='yellow', label= 'Javier')
plt.plot(x4,y4, color='brown', label= 'Juan')
plt.plot(x5,y5, color='red', label= 'Miguel')

plt.title('Popularity of Mexican names in the USA (1880-2020)')
plt.xlabel('Year')
plt.ylabel('Frequency')
plt.legend(loc='upper left');
```

## Limitations
- Female names did not display similar trends most likely due to the resemblance to English names, e.g. Andrea, Angela, Sofia, Patricia, Laura, etc.
- the chosen male names are note exclusive to Mexico
- naming children by immigrants can have unexpected influences such as sentiment, assimilation efforts, etc.
